```python
from seldon_core.seldon_client import SeldonClient
from seldon_core.utils import get_data_from_proto

import matplotlib.pyplot as plt
import numpy as np
```

```python
sc = SeldonClient(
    gateway="ambassador",
    deployment_name="seldon-quantum-dot",
    gateway_endpoint="localhost:8003",
    payload_type="ndarray",
    namespace="seldon",
    transport="rest"
)
```

```python
Bfields = np.linspace(0, 0.2)
prediction = sc.predict(data=Bfields)
```

```python
energies = get_data_from_proto(prediction.response)
```

```python
plt.figure()
plt.plot(Bfields, energies)
plt.xlabel("magnetic field [arbitrary units]")
plt.ylabel("energy [t]")
plt.show()
```
